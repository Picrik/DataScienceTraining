# Import libraries

```python
from afinn import Afinn
import matplotlib.pyplot as plt
import numpy as np
```

# Define the list which will recive our calculations

```python
pos_index = []
neg_index = []
neutral_index = []
```

# Calculate the scores

```python
afinn = Afinn(language="en")
for book in range(1,14):
    with open ("data/" + str(book) + ".txt", "r") as file:
        lines = file.readlines()
        pos = 0
        neg = 0
        neutral = 0
        
        for line in lines:
            score = int(afinn.score(line))
            
            if score > 0:
                pos +=1
            elif score < 0:
                neg+=1
            else:
                neutral+=1
        
        n = len(lines)
        pos_index.append(pos / n)
        neg_index.append(neg / n)
        neutral_index.append(neutral / n)
```

# Plotting the results

```python
X = np.arange(1,14)

plt.plot(X, pos_index, "-.", label="pos")
plt.plot(X, neg_index, "--", label="neg")
plt.plot(X, neutral_index, "-", label="neu")
plt.legend()
plt.xticks(X)
plt.xlabel("Book")
plt.ylabel("Value")
plt.grid()
plt.show()
```
